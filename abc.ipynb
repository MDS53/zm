```python
import pandas as pd
```

```python
df=pd.read_csv('Semi_cleaned_data2.csv')
```

```python
traffic = {
'Low': 0,
'Medium': 1,
'High': 2,
'Jam': 3
}
Vehicle = {
'bicycle':0,
'scooter':1,
'motorcycle':2,
'electric_scooter':3
}
```

```python
df['Road_traffic_density']=df['Road_traffic_density'].replace(traffic)
df['Type_of_vehicle']=df['Type_of_vehicle'].replace(Vehicle)
```
