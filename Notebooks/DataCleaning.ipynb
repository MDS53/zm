# Calculating Distances using lattitudes,Longitudes using Haversine formula

```python
from math import radians, sin, cos, sqrt, atan2
import numpy as np
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points 
    on the earth specified in decimal degrees
    """
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    radius_of_earth = 6371  # Radius of Earth in kilometers
    distance = radius_of_earth * c

    return distance

# Example usage
lat1, lon1 = 12.311072	, 76.654878 # Latitude and longitude of Berlin, Germany
lat2, lon2 = 12.371072	, 76.714878   # Latitude and longitude of Paris, France

distance = haversine(lat1, lon1, lat2, lon2)
print("Distance:", np.round(distance,1), "km")

```

```python
def sub_15_minutes(time_str):
    # Convert the time string to a datetime object
    time = pd.to_datetime(time_str, format='%H:%M')
    # Add 15 minutes
    num=np.random.choice([5,10,15])
    new_time = time - pd.Timedelta(minutes=num)
    # Convert back to time string
    return new_time.strftime('%H:%M')
```
