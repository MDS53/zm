# Data Cleaning

### Cleaned Time Data
- Set TimeData into the correct format. 
- Filled NaN values:
  - **Time_ordered** column: subtracted a random value (5, 10, or 15 minutes) from Time_order_picked values.
  - **Time_order_picked** column: added a random value (5, 10, or 15 minutes) to Time_ordered values.
- Converted **Age** and other relevant columns from object to their appropriate numeric formats.

# Feature Transformation

### Handling Nan Values
- Used KNN Imputer to handle NaN values in features.

### Encoded Categorical Features
- Techniques used:
  - **OnehotEncoding**
  - **OrdinalEncoding**
  - **Frequency encoding**

# Feature Selection

- Dropped unnecessary features which don't affect target variables.

# Feature Construction

- Constructed features based on Time data splitting, calculating distance using lats and longs.
